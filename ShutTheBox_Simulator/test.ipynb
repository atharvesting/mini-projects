```python
# Checking dice result probabilities to design optimal strategy

from game_entities import Player
import matplotlib.pyplot as plt

num_dic = {i+1: 0 for i in range(12)}

p = Player(1)
for i in range(10000):
    num_dic[p.roll_n_dice(2)] += 1

categories = list(num_dic.keys())
frequencies = list(num_dic.values())

plt.figure(figsize=(10, 6))
plt.bar(categories, frequencies, color='skyblue')

# Display the plot
plt.show()
```

```python
def mass_roller(dice_count, roll_count):
    num_dic = {i+1: 0 for i in range(dice_count * 6)}
    p = Player(1)
    for i in range(roll_count):
        num_dic[p.roll_n_dice(dice_count)] += 1

    for key, value in num_dic.items():
        num_dic[key] = value / roll_count

    return num_dic

print(mass_roller(2, 10000000))
```
